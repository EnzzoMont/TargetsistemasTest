```python
# Questão 2
number = float(input("Digite um número: "))

def has_fibonacci(number):
    if number < 0:
        raise ValueError("O número informado é negativo. A sequência de Fibonacci não inclui números menores que 0, por favor digitar um número positivo.")
    fibonacci_sequence = [0, 1]
    while fibonacci_sequence[-1] < number:
        fibonacci_sequence.append(fibonacci_sequence[-1] + fibonacci_sequence[-2])

    if number in fibonacci_sequence:
        return f"O número {number} pertence à sequência de Fibonacci."
    else:
        return f"O número {number} não pertence à sequência de Fibonacci."
has_fibonacci(number)
```

```python
# Questão 3

lower_value = df['valor'].min()
days_lowest_value = df[df['valor'] == lower_value]['dia'].tolist()

print("O menor valor de faturamento:", lower_value)
print("O dias que tiveram o menor valor:", days_lowest_value)
```

```python
# Questão 3

high_value = df['valor'].max()
print("O maior faturamento foi no valor de:", high_value)
```

```python
# Questão 3

mean = df[df['valor'] > 0]['valor'].mean()

days_over_mean = df[df['valor'] > mean]['dia'].count()

print("O total de dias com faturamento acima da média mensal foi:", days_over_mean)
```

```python
# Questão 4

def calculate_percentage(billing_data):

  billing_amount = sum(billing_data.values())

  percentageS = {}
  for state, billing in billing_data.items():
    percentage = (billing / billing_amount) * 100
    percentageS[state] = f"{percentage:.2f}%"

  return percentageS

billing_states = {
  'SP': 67836.43,
  'RJ': 36678.66,
  'MG': 29229.88,
  'ES': 27165.48,
  'Outros': 19849.53
}

percentuais_faturamento = calculate_percentage(billing_states)

for state, percentage in percentuais_faturamento.items():
  print(f"O {state} representou o total de {percentage} do faturamento total.")
```

```python
original_string = input(str("Digite uma frase:"))

def reverse_string(string):

  inverted_string = ""
  for i in range(len(string) - 1, -1, -1):
    inverted_string += string[i]
  return inverted_string

inverted_string = reverse_string(original_string)
print(inverted_string)
```
